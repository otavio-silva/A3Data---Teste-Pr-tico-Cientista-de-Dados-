```python
import os
import numpy
import pandas
import seaborn
from matplotlib import pyplot
```

```python
aeronave = pandas.read_csv(os.path.join("data", "aeronave.csv"), delimiter=';')
```

```python
print(aeronave.columns.to_numpy())
```

```python
aeronave.head()
```

```python
pyplot.figure(figsize=(14, 12))
seaborn.countplot(data=aeronave, x="aeronave_tipo_veiculo", order=aeronave["aeronave_tipo_veiculo"].value_counts().index)
pyplot.show()
```

```python
aeronave.groupby("aeronave_tipo_veiculo").head(10)
```

```python
pyplot.figure(figsize=(14, 12))
seaborn.countplot(data=aeronave, x="aeronave_motor_quantidade", order=aeronave["aeronave_motor_quantidade"].value_counts().index)
pyplot.show()
```

```python
pyplot.figure(figsize=(16, 10))
seaborn.countplot(data=aeronave, y="aeronave_fabricante", order=aeronave["aeronave_fabricante"].value_counts().iloc[:20].index)
pyplot.show()
```

```python
pyplot.figure(figsize=(12, 14))
seaborn.countplot(data=aeronave, y="aeronave_operador_categoria", order=aeronave["aeronave_operador_categoria"].value_counts().index)
pyplot.show()
```

```python
pyplot.figure(figsize=(14, 16))
seaborn.countplot(data=aeronave, y="aeronave_fase_operacao", order=aeronave["aeronave_fase_operacao"].value_counts().index)
pyplot.show()
```

```python
pyplot.figure(figsize=(16, 10))
seaborn.countplot(data=aeronave, x="aeronave_nivel_dano", order=aeronave["aeronave_nivel_dano"].value_counts().index)
pyplot.show()
```

```python
nivel_dano_categoria = aeronave.groupby("aeronave_nivel_dano")["aeronave_registro_categoria"].value_counts().unstack(0).fillna(0)
```

```python
for index in nivel_dano_categoria.index:
    pyplot.figure(figsize=(12, 10))
    nivel_dano_categoria.loc[index].plot.bar()
    pyplot.title(index)
    pyplot.ylabel("Count")
    pyplot.xlabel("Nível de Dano")
    pyplot.show()
```
