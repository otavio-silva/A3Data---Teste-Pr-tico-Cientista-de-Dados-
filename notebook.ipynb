```python
import os
import pandas
import seaborn
from matplotlib import pyplot
```

```python
aeronave = pandas.read_csv(os.path.join("data", "aeronave.csv"), delimiter=';')
```

```python
print(aeronave.columns.to_numpy())
```

```python
aeronave.head()
```

```python
pyplot.figure(figsize=(14, 12))
seaborn.countplot(data=aeronave, x="aeronave_tipo_veiculo", order=aeronave["aeronave_tipo_veiculo"].value_counts().index)
pyplot.show()
```

```python
pyplot.figure(figsize=(14, 12))
seaborn.countplot(data=aeronave, x="aeronave_motor_tipo", order=aeronave["aeronave_motor_tipo"].value_counts().index)
pyplot.show()
```

```python
for motor_tipo in aeronave["aeronave_motor_tipo"].unique():
    pyplot.figure(figsize=(14, 12))
    seaborn.countplot(data=aeronave[aeronave["aeronave_motor_tipo"] == motor_tipo], x="aeronave_nivel_dano", order=aeronave["aeronave_nivel_dano"].value_counts().index)
    pyplot.title(motor_tipo)
    pyplot.show()
```

```python
pyplot.figure(figsize=(14, 12))
seaborn.countplot(data=aeronave, x="aeronave_motor_quantidade", order=aeronave["aeronave_motor_quantidade"].value_counts().index)
pyplot.show()
```

```python
nivel_dano_categoria = aeronave.groupby("aeronave_nivel_dano")["aeronave_registro_categoria"].value_counts().unstack(0).fillna(0)
```

```python
for categoria in aeronave["aeronave_registro_categoria"].unique():
    pyplot.figure(figsize=(16, 10))
    seaborn.countplot(data=aeronave[aeronave["aeronave_registro_categoria"] == categoria].fillna(0), x="aeronave_nivel_dano", order=aeronave.fillna(0)["aeronave_nivel_dano"].value_counts().index)
    pyplot.title(categoria)
    pyplot.show()
```

```python
ocorrencia = pandas.read_csv(os.path.join("data", "ocorrencia.csv"), sep=';')
```

```python
ocorrencia.head()
```

```python
pyplot.figure(figsize=(16, 12))
seaborn.countplot(data=ocorrencia, x="investigacao_status", order=ocorrencia["investigacao_status"].value_counts().index)
pyplot.show()
```

```python
pyplot.figure(figsize=(16, 12))
seaborn.countplot(data=ocorrencia, x="total_aeronaves_envolvidas", order=ocorrencia["total_aeronaves_envolvidas"].value_counts().index)
pyplot.title("Número De Aeronaves Envolvidas")
pyplot.show()
```

```python
years = ocorrencia["ocorrencia_dia"].apply(lambda x: x.split('/')[-1])
pyplot.figure(figsize=(14, 16))
years.value_counts().plot.bar()
pyplot.show()
```

```python
ocorrencia_hora = ocorrencia["ocorrencia_hora"].apply(lambda x: str(x).split(':')[0])
pyplot.figure(figsize=(14, 16))
ocorrencia_hora.value_counts().plot.bar()
pyplot.show()
```

```python
pyplot.figure(figsize=(14, 16))
seaborn.countplot(data=ocorrencia, y="ocorrencia_uf", order=ocorrencia["ocorrencia_uf"].value_counts().index)
pyplot.show()
```

```python
pyplot.figure(figsize=(14, 16))
seaborn.countplot(data=ocorrencia, y="ocorrencia_cidade", order=ocorrencia["ocorrencia_cidade"].value_counts().iloc[:30].index)
pyplot.show()
```

```python
ocorrencia_merge = ocorrencia.merge(ocorrencia_tipo, on="codigo_ocorrencia1", how="left")
```

```python
ocorrencia_merge.head()
```

```python
pyplot.figure(figsize=(16, 10))
seaborn.countplot(data=ocorrencia_merge[ocorrencia_merge["ocorrencia_uf"] == "SP"], x="ocorrencia_classificacao", order=ocorrencia["ocorrencia_classificacao"].value_counts().index)
pyplot.title("Ocorrências - SP")
pyplot.show()
```

```python
pyplot.figure(figsize=(16, 10))
seaborn.countplot(data=ocorrencia_merge[ocorrencia_merge["ocorrencia_uf"] == "MG"], x="ocorrencia_classificacao", order=ocorrencia["ocorrencia_classificacao"].value_counts().index)
pyplot.title("Ocorrências - MG")
pyplot.show()
```

```python
pyplot.figure(figsize=(16, 10))
seaborn.countplot(data=ocorrencia_merge[ocorrencia_merge["ocorrencia_uf"] == "RJ"], x="ocorrencia_classificacao", order=ocorrencia["ocorrencia_classificacao"].value_counts().index)
pyplot.title("Ocorrências - RJ")
pyplot.show()
```
